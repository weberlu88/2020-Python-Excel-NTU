## 找出我們要的縣市，處理資料格式
輸出至excel的forecasts物件格式:
```python
[
    [{day1}, {day2}, {day3}, ...], < Taipei
    [],                            < Taichung
    []                             < Chingchu
]
```
其中每一天的資料是一個字典:
```python
{
    Date: "2/16",
    High: 23,
    Low : 16,
    Description: "多雲時陰短暫陣雨"
}
```

````python
# datetime.datetime.strptime has problems with timezone parsing.
# sample input: '2020-02-16T06:00:00+08:00'
from dateutil import parser # deal with zones, not use
from datetime import datetime

def parse_date(in_str):
    in_str = in_str.split('T')
    date_str = in_str[0]
    time_str = in_str[1].split('+')[0]
    date = datetime.strptime(f"{date_str} {time_str}", '%Y-%m-%d %H:%M:%S')
    return date
````

````python
locs = ['臺北市','臺中市','新竹市']
forecasts = []
data = full_data["result"]["results"]

for loc in locs:
    row = []
    for dataset in data:
        if dataset['locationName'] == loc:
            row.append(dataset)
    forecasts.append(row)
forecasts
````

````python
forecasts[1]
````

````python
# Using pandas
import pandas as pd
df = pd.DataFrame(data)
df2 = df[['parameterName2', 'parameterName3', 'parameterName1', 'locationName', 'startTime']]
df2.rename(columns={'parameterName2':'High',
                    'parameterName3':'Low',
                    'parameterName1':'Description',
                    'locationName'  :'City',
                    'startTime'     :'Date'}, 
           inplace=True)
df2
````

````python
# get 3 city's row
# https://datatofish.com/select-rows-pandas-dataframe/
df_list = []
for city in locs:
    df_list.append(df2.loc[df2['City'] == city])
df_list[1]
````

````python
# parse the 'Date' colomn
# https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#returning-a-view-versus-a-copy

# adj_series = df_list[1]['Date'].apply(lambda s: parse_date(s))
# df_list[1].loc[:, ('Date')] = adj_series

df_list[1].loc[:, ('Date')] = df_list[1]['Date'].apply(lambda s: parse_date(s))
df_list[1]
````

````python
df_list[1].iloc[0]['Date'].date()
````

````python
# 不分白天晚上，只取預報的日期
df_list[1].loc[:, ('Date')] = df_list[1]['Date'].apply(lambda s: s.date())
df_list[1]
````

````python
# 把相同日期的groupby，分別取高溫的max()和低溫的min() 
# 基本(一key) https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html
# 進階(多key) https://stackoverflow.com/questions/25024797/max-and-min-date-in-pandas-groupby
# forecasts = []

df_high = df_list[1].groupby(['Date']).max() #(,as_index = False)
df_low = df_list[1].groupby(['Date']).min()

df_all = df_list[1].groupby(['Date']).agg({'High':max, 
                                           'Low' :min,
                                           'City':'first',
                                           'Description' :'first'})
df_all
````
