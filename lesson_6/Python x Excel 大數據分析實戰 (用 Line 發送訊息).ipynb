# 用 Line 傳訊息

[Line Notify 簡介](https://engineering.linecorp.com/tw/blog/detail/88)


## 1. 首先需要登入 Line Notify 官網

[連結](https://notify-bot.line.me/zh_TW/)


## 2. 產生開發者用的權杖

![](https://www.dropbox.com/s/x7e6k3mqbh4a4nv/token.PNG?dl=1)

登入後，進入個人頁面，在頁面最下方有一個 “發行權杖”

---
## 3. 選擇你想發訊息的社團

![](https://www.dropbox.com/s/escm094rfhhv37x/token2.PNG?dl=1)

建議先選 1 to 1，發送訊息給自己，接下來請記下產生的權杖(一組看似亂碼的字串)

---
## 4. 最後把你的通行碼放入以下程式碼即可

```python
import requests

# 只需要把 content 的內容改成你需要傳遞的訊息即可
content = "你好,我是來自 Python 的訊息，啾咪！"

# Line Notify
line_url = "https://notify-api.line.me/api/notify"
token = "你申請到的 Line 權杖"
headers = {
        "Authorization": "Bearer " + token, 
        "Content-Type" : "application/x-www-form-urlencoded"
    }

payload = {'message': content}
r = requests.post(line_url, headers = headers, params = payload)
```

---

## 5. 用 Line 發送圖片

```python
# Line Notify api 網址
url = "https://notify-api.line.me/api/notify"
# 圖片路徑
pic_url = r"C:\你的圖片路徑\python_logo.png"
# 訊息
msg = "你好,我是來自 Python 的訊息，啾咪！"

headers = {
    "Authorization": "Bearer " + token
}

# 設定訊息内容
payload = { "message": msg }
# 設定圖檔路徑
files = { "imageFile": open(pic_url, "rb") }
# 發送請求
r = requests.post(url, headers = headers, params = payload, files = files)
```
