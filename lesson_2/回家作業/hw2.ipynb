### Calculate 3-day moving average

````python
row = tsmc_sheet.range("A1").end("down").row

# by VBA
def moving_avg(cur_day):
    tsmc_sheet.range(f"C{cur_day}").formula = f"=AVERAGE(B{cur_day-2}:B{cur_day})"
    print( f"C{cur_day}" +" "+ f"=AVERAGE(B{cur_day-2}:B{cur_day})" )
    return

for i in range(4, row+1):
    # get B4 B3 B2's value, calc and write in C4
    moving_avg(i)
````

### Judge whether to buy or to sale, and calculate the *cash/stock on hand*, *total asset*.

````python
# vars
start_cash = 1000000
cash = 1000000 
shares = 0
assets = cash
unit = 1000
revenue = 0

cls_price = 0
````

````python
# BUY it if the 'closing market price' >= 'moving avgerage', and have enough cash to purchase.
def buy_ornot(day):
    cls_price = tsmc_sheet.range(f"B{day}").value
    mv_price  = tsmc_sheet.range(f"C{day}").value
    if (cls_price >= mv_price) and (cash >= unit*cls_price):
        return True
    else:
        return False

def do_nothing(day):
    # write purchase amount
    tsmc_sheet.range(f"D{day}").value = 0
    tsmc_sheet.range(f"E{day}").value = 0
    
    # write cash/stock on hand & assets
    tsmc_sheet.range(f"F{day}").value = shares 
    tsmc_sheet.range(f"G{day}").value = cash 
    tsmc_sheet.range(f"H{day}").value = assets 
    return

# Purchase an unit of stock
def do_purchase(day):
    global cls_price, cash, shares, assets, revenue, start_cash
    cls_price = get_cls_price(day)
    
    # operations in background
    shares += unit
    cash   -= unit * cls_price
    assets  = recount_assets()
    revenue = assets - start_cash
    
    # write purchase amount
    tsmc_sheet.range(f"D{day}").value = unit 
    tsmc_sheet.range(f"E{day}").value = 0
    
    # write cash/stock on hand & assets
    tsmc_sheet.range(f"F{day}").value = shares 
    tsmc_sheet.range(f"G{day}").value = cash 
    tsmc_sheet.range(f"H{day}").value = assets 
    
    # write revenue
    tsmc_sheet.range("K15").value = revenue
    return

# Sale an unit of stock
def do_sale(day):
    global cls_price, cash, shares, assets, revenue, start_cash
    cls_price = get_cls_price(day)
    
    # operations in background
    shares -= unit
    cash   += unit * cls_price
    assets  = recount_assets()
    revenue = assets - start_cash
    
    # write purchase amount
    tsmc_sheet.range(f"D{day}").value = 0 
    tsmc_sheet.range(f"E{day}").value = unit
    
    # write cash/stock on hand & assets
    tsmc_sheet.range(f"F{day}").value = shares 
    tsmc_sheet.range(f"G{day}").value = cash 
    tsmc_sheet.range(f"H{day}").value = assets 
    
    # write revenue
    tsmc_sheet.range("K15").value = revenue
    return

def get_cls_price(day):
    return tsmc_sheet.range(f"B{day}").value

def recount_assets():
    share_market_value = shares * cls_price
    return cash + share_market_value
````

````python
# On the fourth day, buy it first.
cls_price = get_cls_price(4)
shares += unit
cash   -= unit * cls_price
assets  = recount_assets() 

tsmc_sheet.range("D4").value = unit # purchase amount
tsmc_sheet.range("E4").value = 0
tsmc_sheet.range("F4").value = shares # stock on hand
tsmc_sheet.range("G4").value = cash # cash on hand
tsmc_sheet.range("H4").value = assets # assets

print("cls_price: ", cls_price, ", shares: ",shares)
````

````python
# next N-days after.
for i in range(5, row+1):
    if buy_ornot(i) is True:
        do_purchase(i)
        print(i, "Buy.  shares: ", shares, " rev: ", revenue)
    elif shares >= unit:
        do_sale(i)
        print(i, "Sale. shares: ", shares, " rev: ", revenue)
    else:
        do_nothing(i)
        print(i, "Do nothing. shares: ", shares)
````
